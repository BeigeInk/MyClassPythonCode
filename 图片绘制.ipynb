```python
from sklearn.datasets import load_iris
import numpy as np
import pandas as pd
```

```python
iris=load_iris()
```

```python
iris_pd=pd.DataFrame(data=iris.data,columns=iris.feature_names)
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=2)
pca=pca.fit(iris.data)
t_d=pca.transform(iris.data)
```

```python
y=iris.target
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure()
plt.scatter(t_d[y==0,0],t_d[y==0,1],c='red',label=iris.target_names[0])
plt.scatter(t_d[y==1,0],t_d[y==1,1],c='green',label=iris.target_names[1])
plt.scatter(t_d[y==2,0],t_d[y==2,1],c='blue',label=iris.target_names[2])
plt.legend()
plt.title('two feature after d-redution')
```

```python
fig,axes=plt.subplots(2,3)
#help(axes[0][0])
r_dir=[[0,1],[0,2],[0,3],[1,2],[1,3],[2,3]]
for out_r in [0,1]:
    for in_r in [0,1,2]:
        in_dir=out_r*3+in_r
        axes[out_r][in_r].scatter(iris.data[y==0,r_dir[in_dir][0]],iris.data[y==0,r_dir[in_dir][1]],c='red',label=iris.target_names[0])
        axes[out_r][in_r].scatter(iris.data[y==1,r_dir[in_dir][0]],iris.data[y==1,r_dir[in_dir][1]],c='green',label=iris.target_names[1])
        axes[out_r][in_r].scatter(iris.data[y==2,r_dir[in_dir][0]],iris.data[y==2,r_dir[in_dir][1]],c='blue',label=iris.target_names[2])
       # axes[out_r][in_r].legend()
       # axes[out_r][in_r].subtitle('two feature after d-redution')
```
